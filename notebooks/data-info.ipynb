# Informações sobre o dataset

```python
import os
import pandas as pd
```

```python
# Cria um objeto dataset, para armazenar as informações necessárias
class Dataset:
    def __init__(self, name: str):
        self.name = name.lower()
        self.path = os.path.join("..", "data", "02-cleaned", self.name + ".csv")
        self.df = pd.read_csv(self.path)

        print(f"Dataset '{name}' criado com sucesso, com {self.df.shape[0]} registros)!")

    def info(self):
        print(f"\n##### {self.name}")
        print(f"# Caminho: \'{self.path}\'")

        # Verifica quantos são toxic e non-toxic
        qtd_toxic = len(self.df[self.df['is_toxic'] == True])
        qtd_non_toxic = len(self.df[self.df['is_toxic'] == False])
        print(f"# Quantidade de toxic: {qtd_toxic} ({qtd_toxic / len(self.df) * 100:.2f}%)")
        print(f"# Quantidade de non-toxic: {qtd_non_toxic} ({qtd_non_toxic / len(self.df) * 100:.2f}%)")
```
